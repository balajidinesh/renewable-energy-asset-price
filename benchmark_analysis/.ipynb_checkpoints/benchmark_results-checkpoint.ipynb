```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the models data is  from regression resulst can be found inside folder names regression in smae directory
models = [ 'LightGBM', 'SVR', 'XGBoost', 'RF', 'CatBoost' , 'CNN-LSTM']

# Precise MAE values for SENT and NO_SENT
mae_sent = [ 0.6708455911042006, 0.6636867346443627, 0.6433188066637815, 0.641030917504331, 0.6390009182332387,0.5605123780101263]
mae_no_sent= [ 0.6786234124843215, 0.6640878467187996, 0.6398635041130425, 0.6480842220108459, 0.6377435343592206 ,0.6303725175287177]

# Precise DA values for SENT and NO_SENT
da_sent = [0.56448, 0.5489955555555556, 0.5663288888888889, 0.5795022222222223, 0.57904 , 0.6786666666666666]
da_no_sent = [ 0.56448, 0.5443733333333334, 0.5672533333333334, 0.5792711111111111, 0.5758044444444445 , 0.64]


figsize = (6,4)
alim = np.array([0.45, 0.7])
blim = np.array([0.45, 0.7]) * 100
bbox_to_anchor=(0.5, -0.2)
ffig1, ax1 = plt.subplots(figsize=figsize)

# Horizontal bar chart for MAE comparison
bar_width = 0.25
index = np.arange(len(models))

lsent = "With Sentiment"
lnosent = "Without Sentiment"
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 10


ax1.barh(index , mae_no_sent, bar_width, label=lnosent, color='lightgreen')
ax1.barh(index+ bar_width, mae_sent, bar_width, label=lsent, color='skyblue')

# ax1.set_ylabel('Models')  # Change to ylabel since bars are horizontal
ax1.set_xlabel('Average MAE')     # Adjust label for horizontal bar orientation
ax1.set_title('MAE Comparison across Models')
ax1.set_yticks(index + bar_width / 2)
ax1.set_yticklabels(models)
plt.legend(loc='upper center', bbox_to_anchor=bbox_to_anchor, ncol=2,frameon=False)
ax1.set_xlim(alim)  # Adjust limit for horizontal axis (x-axis)
# Adjust layout and display the MAE plot
plt.tight_layout()
plt.savefig("mae_comparison.png", format='png', dpi=300)
plt.show()
  # Save as PNG with high resolution

# Second plot: DA Comparison
fig2, ax2 = plt.subplots(figsize=figsize)

# Horizontal bar chart for DA comparison
ax2.barh(index , np.array(da_no_sent)*100, bar_width, label=lnosent, color='orange')
ax2.barh(index+ bar_width, np.array(da_sent)*100, bar_width, label=lsent, color='salmon')

# ax2.set_ylabel('Models')  # Change to ylabel for horizontal orientation
ax2.set_xlabel('Average Directional Accuracy (DA %)')
ax2.set_title('DA Comparison across Models')
ax2.set_yticks(index + bar_width / 2)
ax2.set_yticklabels(models)
plt.legend(loc='upper center', bbox_to_anchor=bbox_to_anchor, ncol=2,frameon=False)
ax2.set_xlim(blim)

# Adjust layout and display the DA plot
plt.tight_layout()
plt.savefig("da_comparison.png", format='png', dpi=300) 
plt.show()

```
