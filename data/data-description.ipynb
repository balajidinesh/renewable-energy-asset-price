```python
import pandas as pd
import matplotlib.pyplot as plt

# List of files and column name
files = ['ICLN', 'PBD', 'QCLN']

# Dictionary to store the results for each file

def gen_des_stats(col_name='log_return'):
    results = {}
    
    for file in files:
        # Construct the file path
        file_path = f"{file}/{file}_INPUT.csv"
        
        # Read the CSV file
        df = pd.read_csv(file_path)
        
        # Select the column based on col_name
        if col_name in df.columns:
            data = df[col_name].dropna()  # Drop NaN values
            
            # Calculate the statistics
            stats = {
                'Mean': data.mean(),
                'Maximum': data.max(),
                'Minimum': data.min(),
                'Standard Deviation': data.std()
            }
            
            # Store the results for the current file
            results[file] = stats
            
            # Generate histogram with bin size 50
            plt.figure(figsize=(10, 6))
            plt.hist(data, bins=50, edgecolor='black')
            plt.title(f'Histogram of {col_name} for {file}')
            plt.xlabel(col_name)
            plt.ylabel('Frequency')
            plt.grid(True)
            
            # Save the histogram as a file
            plt.savefig(f'{file}_histogram.png')
            plt.show()

    return pd.DataFrame(results)



```

```python
# Display the calculated statistics
stats_df = gen_des_stats(col_name='log_return')
print(stats_df)
```
